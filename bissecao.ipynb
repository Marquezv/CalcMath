### Utilizando o metodo da Bisseção encontre a raiz da função

## $ f(x) = e^x - x - 2 $

```python
import numpy as np
import matplotlib.pyplot as plt
from math import e
```

```python
def f(x):
    return e**x - x - 2

a = -2
b = 0
erro = 1e-3

x_values = np.linspace(a, b, 1000)
y_values = [f(x) for x in x_values]

n_iter = 0
a_values = []
b_values = []

while (b - a) / 2 > erro :
    c = (a + b) / 2
    a_values.append(a)
    b_values.append(b)

    if f(c) == 0:
        break
    elif f(a) * f(c) < 0:
        b = c
    else:
        a = c
    n_iter += 1

print(f"Raiz aproximada: {c} após {n_iter} iterações")
```

```python
# Gráfico do processo
plt.plot(x_values, y_values, label='f(x)')
plt.plot(a_values, [f(a) for a in a_values], 'ro', label='a values')
plt.plot(b_values, [f(b) for b in b_values], 'go', label='b values')
plt.axhline(0, color='red', linestyle='--', linewidth=0.7)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Processo da Bisseção')
plt.legend()
plt.grid(True)
plt.show()
```
